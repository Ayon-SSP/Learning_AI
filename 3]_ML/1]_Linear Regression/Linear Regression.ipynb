```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
m = reg.coef_
b = reg.intercept_
print(reg.coef_)
print(reg.intercept_)
```

```python
y = m*(3300) + b
y
```

```python
# Time to predict the price of a house with all the given areas form area.csv
area_df = pd.read_csv("areas.csv")
area_df.head(3)
```

```python
p = reg.predict(area_df)
```

```python
area_df['price'] = p
area_df.to_csv('prediction.csv',index=False)
area_df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```

```python
# Model is saved in the reg 
import pickle
with open('model_pickle','wb') as f:
    pickle.dump(reg,f)
```

```python
# Model is saved in the reg to retrive it
with open('model_pickle','rb') as f:
    model = pickle.load(f)
```

```python
model.predict([[3300]])
```

```python
from sklearn.externals import joblib
joblib.dump(model,'model_joblib')
mj = joblib.load('model_joblib')
mj.pridict(5000)
```
