```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("bangalore house price prediction cleaned.csv")
df
```

```python
ques = df.drop('price',axis=1)
ans = df['price']
# df.drop('price',axis=1)
ques
```

```python
reg = linear_model.LinearRegression()
reg.fit(ques,ans)
```

```python
ques[:5][:]
```

```python
ans[:5]
```

```python
predectedvals = reg.predict(ques[:5][:])
```

```python
# compare the predicted price with the actual price
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(predectedvals, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
# Time to predict the price of a house with all the given areas form area.csv
area_df = pd.read_csv("areas.csv")
area_df.head(3)
```

```python
p = reg.predict(area_df)
```

```python
area_df['price'] = p
area_df.to_csv('prediction.csv',index=False)
area_df
```

```python
# Model is saved in the reg 
import pickle
with open('model_pickle','wb') as f:
    pickle.dump(reg,f)
```
